# Сборный проект

Вы работаете в интернет-магазине «Стримчик», который продаёт по всему миру компьютерные игры. Из открытых источников доступны исторические данные о продажах игр, оценки пользователей и экспертов, жанры и платформы (например, Xbox или PlayStation). ESRB оценивает игровой контент и присваивает ему подходящую возрастную категорию, например, «Для взрослых», «Для детей младшего возраста» или «Для подростков».

# Подготовка данных

Приведем названия колонок к нижнему регистру

````python
df.columns = df.columns.str.lower()
df.columns
````

````python
df['year_of_release'] = pd.to_datetime(df['year_of_release'], format='%Y').dt.year
````

````python
df.query('user_score == "tbd"')
````

Переведем оценки критиков и пользователей в int

````python
#попытки перевести данные в int с игнорированием ошибки об NaN не удались, пришлось для начала заполнять NaN
df['critic_score'] = pd.to_numeric(df['critic_score'], errors='coerce').astype(int, errors='ignore')
df['user_score'] = pd.to_numeric(df['user_score'], errors='coerce').astype(int, errors='ignore')
df.info()
````

* После проверки нескольких тайтлов стало ясно, что оценки были взять с сайта metacritic.com. Оценки игроков могут отличаться от текущих оценок на сайте, т.к. пользовательские рецензии публикуются не только на старте продаж, как в случае с оценкой критиков, поэтому со временем средняя оценка игроков может меняться. Из-за того, что сайт запустился в 2000 году, у многих старых игр отсутствуют оценки.<br>

* Отсутствующие оценки игроков и критиков вряд ли стоит заполнять медианой по платформе, жанру или году - во-первых, оценки отсутствуют у слишком большого числа игр и заполнение их средним может сильно исказить реальную картину, во-вторых, оценки игр слишком непредсказуемы, и здесь нельзя с уверенностью сказать, что даже в рамках одной долгой и успешной франшизы они стабильны (пример - Call of Duty, которая имеет достаточно большой разброс оценок). Поэтому безопаснее их заполнить маркером, например -1.<br>
* В столбцах с пользовательским рейтингом попадается достаточно много значений 'tbd'. Оно расшифровывается как 'To be determined' и значит, что ещё не набралось достаточно рецензий для вывода общего рейтинга. При переводе в int оно распознаётся как NaN и заменяется на маркер<br> 

````python
df['critic_score'] = df['critic_score'].fillna(-1).astype('int64')
#домножим на 10, чтобы оценки были в 100-балльной системе, как у critic_score
df['user_score'] = df['user_score'].apply(lambda x: x * 10).fillna(-1).astype('int64')
df.info()
````

Половина рейтингов представлена в очень малом объёме. Часть из них уже устарела (K-A это нынешний E), часть - можно объединить с более крупными близкими группами. Например, EC - это для для детей от 3 лет и в целом его также можно слить вместе с E, а AO (Adults Only) с Mature. RP - Rating Pending, т.е. выставление рейтинга ещё ожидается. Его заменим на 'undefined', как и остальные пропуски - на медиану в заменах тут полагаться также нельзя

````python
df['rating'].value_counts()
````

````python
def common_rating(row):
    esrb = {'E10+': 'Everyone 10+', 'E': 'Everyone', 'T': 'Teen', 'M': 'Mature', 
                 'AO': 'Mature', 'RP': 'Undefined', 'EC': 'Everyone', 'K-A': 'Everyone'}
    row['rating'] = esrb[row['rating']]
    return row['rating']
````

````python
df['rating'] = df[df['rating'].notnull()].apply(common_rating, axis = 1)
````

````python
df['rating'] = df['rating'].fillna('Undefined')
````

````python
df.isna().sum()
````

В жанре, названии игры всего по 2 пропуска, в годе выпуска - 269 пропусков, что 1.5% данных. Учитывая, что основные подсчёты будут делаться по годам, строки с пустым годом нам не помогут, при таком количестве их можно удалить

````python
df = df.dropna(subset=['name', 'genre', 'year_of_release'])
````

````python
df['year_of_release'] = df['year_of_release'].astype('int64')
````

````python
def vizualising(df, index_col, values_col, func, kind, figsize, title1):
    vizualisation = (df.pivot_table(index=index_col, values=values_col, aggfunc=func)
                     .plot(kind=kind, figsize=figsize, grid=True, title=title1))
    return vizualisation
````

* Построим диаграмму рассеяния для оценок и продаж PS4 и Xbox One

````python
fig, ax = plt.subplots(1,4)
fig.set_size_inches(20, 5)
num_of_subplot = 0
for console in ['PS4', 'XOne']:
    for score in ['critic_score', 'user_score']:
        (ps4_xone[ps4_xone[score] != -1].plot(kind='scatter', y='total_sales', x=score, ax = ax[num_of_subplot], 
                                              title=console, alpha=0.2))
        num_of_subplot += 1
````

````python
for console, score in zip(['PS4', 'PS4', 'XOne', 'XOne'], ['critic_score', 'user_score', 'critic_score', 'user_score']):
    print(f'Корреляция между total_sales и {score} для {console}:', 
          (ps4_xone[(ps4_xone[score] != -1) & (ps4_xone['platform'] == console)])[score]
          .corr(ps4_xone[(ps4_xone[score] != -1) & (ps4_xone['platform'] == console)]['total_sales']))
````

По шкале Чеддока у нас наблюдается связь между оценками критиков и продажами для обеих платформ, слабая междц оценками игроков и продажами на PS4 и отсутствует связь для продаж на Xbox One. Возможно это из-за того, что основные продажи делаются в первые дни после релиза, когда игроки ориентируются только на оценки критиков, которые появляются взаранее. Оценки игроков обычно появляются позднее и бывают более полярные и субъективные, на них ориентируются в меньшую степень.

````python
correlation_sales_users = []
correlation_sales_critics = []

for console in actual_df['platform'].unique():
    for score in ['critic_score', 'user_score']:
        corr = (actual_df[(actual_df[score] != -1) & (actual_df['platform'] == console)][score]
                .corr(actual_df[(actual_df[score] != -1) & (actual_df['platform'] == console)]['total_sales']))
        if score == 'critic_score':
            correlation_sales_critics.append(corr)
        else:
            correlation_sales_users.append(corr)
````

# Проверка гипотез

Проверим средние пользовательские рейтинги платформ Xbox One и PC на равенство. Пусть нулевой гипотезой будет то, что они равны, а альтернативной - что различны. p-value будет равен 0.05. Для проверок гипотезы о равенстве среднего двух генеральных совокупностей по взятым из них выборкам используем t-тест. Так как генеральные совокупности у нас с разными параментрами (пользователи, которые оценили игры на Xbox One, и те, что оценили на PC), то параметр equal_var = False

````python
from scipy import stats as st

xone_user_rating = actual_df.query('platform == "XOne" and user_score != -1')['user_score']
pc_user_rating = actual_df.query('platform == "PC" and user_score != -1')['user_score']
````

````python
xone_user_rating.var()
````

````python
pc_user_rating.var()
````

````python
xone_user_rating.mean()
````

````python
pc_user_rating.mean()
````

````python
results = st.ttest_ind(xone_user_rating, pc_user_rating, equal_var = False)

print('p-значение: ', results.pvalue)
alpha = 0.05
if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else: 
    print("Нулевую гипотезу отвергнуть нельзя")
````

Опровергнуть гипотезу не удалось, игры на PC и Xbox One имеют схожий средний рейтинг

Теперь проверим на равенство средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования»). Нулевой гипотезой будет их одинаковость, альтернативной - их различие. Здесь также проверка гипотезы о равенстве среднего двух генеральных совокупностей по взятым из них выборкам. Дисперсия выборок сильно разнится, поэтому ставим параметр equal_var = False

````python
sports_user_rating = actual_df.query('genre == "Sports" and user_score != -1')['user_score']
action_user_rating = actual_df.query('genre == "Action" and user_score != -1')['user_score']
````

````python
sports_user_rating.var()
````

````python
action_user_rating.var()
````

````python
sports_user_rating.mean()
````

````python
action_user_rating.mean()
````

````python
results = st.ttest_ind(action_user_rating, sports_user_rating, equal_var = False)

print('p-значение: ', results.pvalue)
alpha = 0.05
if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else: 
    print("Нулевую гипотезу отвергнуть нельзя")
````

## Вывод

Мы проверили две гипотезы, о равенстве пользовательских оценок для PC и XOne, экшенов и спортсимов. В первом случае нулевая гипотеза устояла, во втором была статистически опровергнута.

* По сравнению с другими платформами, у PS4 и Xbox One самый большой верхний квартиль и максимальное значение. Консоли предыдущего поколения (PS3, Xbox 360, Wii) ещё показывают высокие продажи игр, но уже уступают набирающим силу новичкам.<br>
* По шкале Чеддока у нас наблюдается связь между оценками критиков и продажами для PS4 и Xbox One платформ, слабая между оценками игроков и продажами на PS4 и отсутствует связь для продаж на Xbox One. Возможно это из-за того, что основные продажи делаются в первые дни после релиза, когда игроки ориентируются только на оценки критиков, которые появляются взаранее. Оценки игроков обычно появляются позднее и бывают более полярные и субъективные, на них ориентируются в меньшую степень. Имеет смысл при продвижении игр упоминать их средний рейтинг на Metacritic и приводить цитаты из обзоров авторитетных изданий - это должно увеличить продажи.<br>
* Упор стоит делать на игры в жанре экшен, шутер, РПГ и спортивный симулятор, если говорить о продажах в Европе и Америке. В игры этих жанров удобнее всего играть на консолях и они популярны в этих регионах уже давно.<br>
* В Японии отличается от других рынков - здесь доминирует портатив и отечественные производители (но в основном Nintendo). Здесь стоит сделать акцент на играх для 3DS, PS4, WiiU, на первый план выдвинуть ролевые игры, экшены, музыкальные игры и файтинги.<br>
* На Западе можно одинаково успешно продвигать игры и для взрослой аудитории, и для семейной, на японском рынке лучше сосредоточиться на семейных играх рейтинга E и E10+.
